```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
yasList=[10,20,30,40,50,60,70,75]
```

```python
kiloList=[20,60,80,85,86,87,90,75]
```

```python
numpyYasList=np.array(yasList)
numpyKiloList=np.array(kiloList)
```

```python
plt.plot(numpyYasList,numpyKiloList,"g")
plt.xlabel("yas")
plt.ylabel("Kilo")
plt.title("Kilonun yaşa Göre Değişim Grafiği")
```
