```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
x=["1.ABD","2.ÇİN","3.JAPONYA","4.ALMANYA","5.İNGİLTERE","6.FRANSA"]
y=[20410,14090,5170,4210,2940,2930]
z=[63039,9957,40649,51158,44162,44915]
```

```python
plt.plot(x,y,"r")
plt.xlabel("ÜLKELER")
plt.ylabel("MİLLİ GELİR")
plt.title("ÜLKELERE GÖRE MİLLİ GELİR")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,"r--")
plt.subplot(1,2,2)
plt.plot(x,z,"y*-")
plt.tight_layout()
```

```python
f=plt.figure()
axes=f.add_axes([0.1,0.1,0.9,0.9])
axes.plot(x,y)
axes.set_xlabel("ÜLKELER")
axes.set_ylabel("MİLLİ GELİR")
axes.set_title("ÜLKELERE GÖRE MİLLİ GELİR")
```

```python
f=plt.figure()
axes1=f.add_axes([0.1,0.1,0.9,0.9])
axes2=f.add_axes([0.5,0.5,0.4,0.4])
axes1.plot(x,y,"y")
axes2.plot(x,z,"r")
```

```python
fig,axes=plt.subplots(nrows=1, ncols=2)

axes[0].plot(x,y,"b")
axes[0].set_title("MİLLİ GELİR")

axes[1].plot(x,z,"y")
axes[1].set_title("KİŞİ BAŞINA")

plt.tight_layout()
```

```python
f=plt.figure()
axes=f.add_axes([0.1,0.1,0.9,0.9])
axes.plot(x,y,color="red",linewidth=1,linestyle="-",marker="o",markersize=20,markerfacecolor="yellow",markeredgewidth=4,markeredgecolor="blue",alpha=0.9)
axes.set_xlabel("ÜLKELER")
axes.set_ylabel("MİLLİ GELİR")
axes.set_title("ÜLKELERE GÖRE MİLLİ GELİR")
```

```python
f=plt.figure()
axes=f.add_axes([0.1,0.1,0.9,0.9])
axes.plot(x,y,color="red",linewidth=1,linestyle="-")
axes.set_xlabel("ÜLKELER")
axes.set_ylabel("MİLLİ GELİR")
axes.set_title("ÜLKELERE GÖRE MİLLİ GELİR")

axes.set_xlim([2,6])
axes.set_ylim([2000,7500])
```

```python
f=plt.figure(figsize=(10,6))
axes=f.add_axes([0,0,1,1])
axes.plot(x,y)
axes.set_xlabel("ÜLKELER")
axes.set_ylabel("MİLLİ GELİR")
axes.set_title("ÜLKELERE GÖRE MİLLİ GELİR")
```

```python
fig,axes=plt.subplots(nrows=1, ncols=2,figsize=(13,5))

axes[0].plot(x,y,"b")
axes[0].set_title("MİLLİ GELİR")

axes[1].plot(x,z,"y")
axes[1].set_title("KİŞİ BAŞINA")

plt.tight_layout()
```

# piecharts

```python
labels = 'TÜRKİYE', 'ABD', 'ÇİN', 'ALMANYA'
sizes = [15, 35, 35, 15]
explode = (0.2, 0, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
```
